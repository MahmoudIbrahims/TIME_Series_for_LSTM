```python
df=pd.read_excel('data.xlsx', engine='openpyxl')
pd.options.display.max_rows = 10000
df.head(3)
```

#### Add an extra column for "Order Month" & "Order Year"

```python
df['Order Date year']=df['Order Date'].dt.year
df['Order Date month']=df['Order Date'].dt.month
```

```python
df.head()
```

#### Add a new column to calcolate the profit margin for each sales record  

```python
df['profit margin']=df['Profit']/df['Sales']
df.head()
```

#### Export manipulated Dataframe to csv

```python
df.to_csv('E:\DATA_output.csv', index=False)
```

#### create a new dataframe to reflect total  by sub_category

```python
# Group By Sub-Category [SUM]
df_by_sub_category = df.groupby('Sub-Category').sum()

# Reset Index
df_by_sub_category.reset_index(inplace=True)
```

```python
df_by_sub_category.head()
```

#### Develop a function,to return a dataframe which is grouped by a particular column

```python
def grouped_data(column_name):
    df_tmp=df.groupby(column_name).sum()
    df_tmp.reset_index(inplace=True)
    return df_tmp
```

```python
grouped_data('Sales').head(10)
```

```python
grouped_data('Order Date year')
```

**Show the distribution and skewness of Sales and Order Date month [scatter]**
